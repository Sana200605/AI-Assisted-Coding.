```python
import numpy as np
math_score=np.array(df['math_score'])
```

```python
print(math_score)
```

```python
import numpy as np
np.mean
```

```python
import numpy as np
np.median
```

```python
mean_math_score = np.mean(math_score)
print(f"The mean math score is: {mean_math_score}")
```

```python
science_score = np.array(df['science_score'])
mean_science_score = np.mean(science_score)
print(f"The mean science score is: {mean_science_score}")
```

```python
english_score = np.array(df['english_score'])
mean_english_score = np.mean(english_score)
print(f"The mean english score is: {mean_english_score}")
```

```python
median_math_score = np.median(math_score)
print(f"The median math score is: {median_math_score}")
```

```python
median_science_score = np.median(science_score)
print(f"The median science score is: {median_science_score}")
```

```python
median_english_score = np.median(english_score)
print(f"The median english score is: {median_english_score}")
```

```python
import matplotlib.pyplot as plt
plt.hist(math_score)
plt.title('Distribution of Math Scores')
plt.xlabel('Math Score')
plt.ylabel('Frequency')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Gather the mean scores
average_scores = {
    'Math': mean_math_score,
    'Science': mean_science_score,
    'English': mean_english_score
}

# Create a pandas Series from the dictionary
average_scores_series = pd.Series(average_scores)

# Create the bar chart
plt.figure(figsize=(8, 6))
average_scores_series.plot(kind='bar', color=['skyblue', 'lightcoral', 'lightgreen'])
plt.title('Average Scores Across Subjects')
plt.xlabel('Subject')
plt.ylabel('Average Score')
plt.ylim(0, 100) # Assuming scores are out of 100
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
